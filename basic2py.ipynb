```python
#nested while loops
n = int(input("enter the no of times"))
 
i = 1
while i <= n :
  j = 1
  while j <= i:
    print("*",end="")
    j = j+1

  print()
  i = i+1


```

```python
def ask_yes_no(prompt):
  while True:
    answer = input(prompt ='(y or n)')
    if answer == 'y' or answer =='Y':
      return True
    if answer == 'N ' or 'N':
      return False
    print("answer Y or N")
# print y or no
if ask_yes_no("do you like pizza?"):
    print("you like pizza")
else:
    print("you dont like pizza")
```
