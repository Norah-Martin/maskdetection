```python
#used opencv for displaying the detected images (which i detect in above cell)
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread("/content/yolov5/runs/detect/exp2/groupimg1.jpg")
height, width = image.shape[:2]
resized_image = cv2.resize(image,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)

fig = plt.gcf()
fig.set_size_inches(10, 10)
plt.axis("off")
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.show()
```
