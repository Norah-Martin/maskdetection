```python
#edge direction techiniques
#sobel
#laplacian
#canny
```

```python
#SOBEL
img =cv2.imread('/content/sudoku-image.png')
sobel_x =cv2.Sobel(img,cv2.CV_8U,dx=1,dy=0,ksize =-1)
sobel_y =cv2.Sobel(img,cv2.CV_8U,dx=0,dy=1,ksize =-1)
sobel = cv2.bitwise_or(sobel_x,sobel_y)
cv2_imshow(sobel_x)
```

```python
cv2_imshow(sobel_y)
```

```python
cv2_imshow(sobel)
```

```python
#LAPLACIAN

laplacian =cv2.Laplacian(img,cv2.CV_8U)
cv2_imshow(laplacian)
```

```python
#CANNY edge detection
img = cv2.imread('/content/sudoku-image.png',0)
canny = cv2.Canny(img,150,100)
cv2_imshow(canny)
```

```python
#HAARCASCADE CLASSIFIER
#facial recognition
```

```python
face_cascade =cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('/content/aaa.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

face =  face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5)
for x,y,w,h in face:
  img =cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),7)

cv2_imshow(img)
```
