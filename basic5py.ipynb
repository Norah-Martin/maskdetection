```python
#LINE GRAPH USING LISTS
import matplotlib.pyplot as plt
a = [1,2,3,4,5]
b = [7,8,9,10,3]
c =(1,3,4,5,6) #tuple
plt.plot(a,b,label="line1",linestyle="dashed",linewidth=8) #default colours=blue orange green black
plt.plot(a,c,label="line2",color="red",marker="*",linestyle ="dashed")
plt.plot(b,c,label ="line 3",marker="*",markersize=10,linestyle = "dashdot")
plt.title(" LINE GRAPH")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()#line 1 line 2 in side is legends
plt.show()
```

```python
#BAR GRAPH
name = ["a","b","c"]
height=[150,160,170]
plt.bar(name,height,color=["r","g","b"])
plt.show()
```

```python
ax.set_facecolor("black")
name = ["a","b","c"]
height=[150,160,170]
plt.barh(name,height,color=["r","g","b"]) #barh -horizontal
plt.show()
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt

hours =["9","8","2","4"]
work = ["colege","study","food","fun"]
cols1 = ["r","g","pink","b"]
plt.pie(hours,labels = work,colors=cols1,startangle =90,autopct ="%1.1f%%",explode=[0,0,1,0]) 
plt.show()
```

```python
#scater plot
a =[1,2,3,4,5]
b = [6,7,8,9,1]
plt.scatter(a,b,color='r',s=100,marker="*",alpha=0.9)#alpha transparency
 plt.show()
```

```python
tips = sns.load_dataset('tips')
tips #prints a dataframe
```

```python
sns.set()#sets all the default styles for the dataframe
```

```python
sns.distplot(tips['total_bill'])
#hist-histogram
#kde - kernel density estimate
```

```python
sns.distplot(tips['total_bill'],kde=False)#only histogram
```

```python
sns.distplot(tips['total_bill'],hist=False)
```

```python
sns.scatterplot(x="total_bill",y="tip",data= tips,color='red',hue='day')
```

```python
sns.relplot(x="total_bill",y='day',data=tips)
```

```python
#joint plot
sns.jointplot(x="total_bill",y="tip",data=tips,kind="resid",color='r') #resid plot regression y on x
sns.jointplot(x="total_bill",y="tip",data=tips,kind="scatter")
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg",color='k')
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde",color='g',alpha=1)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex",color='r')
```

```python
sns.pairplot(tips)
```

```python
sns.stripplot(y="total_bill",x="day",data=tips)
```

```python
sns.swarmplot(y="total_bill",x="day",data=tips)
```

```python
sns.boxplot(y="total_bill",x="day",data=tips)
```

```python
sns.boxplot(x=tips["total_bill"])
```

```python
sns.violinplot(y="total_bill",x="day",data=tips,color='g')
```

```python
#categorical plot
sns.catplot(y="total_bill",x="day",data=tips) #stripplot
sns.catplot(y="total_bill",x="day",data=tips,kind='box')
sns.catplot(y="total_bill",x="day",data=tips,kind='boxen')
sns.catplot(y="total_bill",x="day",data=tips,kind='violin')
```
